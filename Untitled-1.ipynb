```python
import numpy as np

def compute_hrv_index(ibis):
    # Replace this with your HRV index computation logic
    # For example, you can use time-domain or frequency-domain methods
    # Here, I'm using the standard deviation as a simple example
    hrv_index = np.std(ibis)
    return hrv_index

def process_ibis(input_file):
    # Load data from the file
    data = np.loadtxt(input_file)

    # Extract time and IBI columns
    time_col = data[:, 0]
    ibi_col = data[:, 1]

    # Set the window size (5 minutes)
    window_size_seconds = 5 * 60

    # Initialize arrays
    ibi_array = []
    hrv_indices = []

    for i in range(len(time_col)):
        # Append the current IBI to the array
        ibi_array.append(ibi_col[i])

        # Remove the oldest IBIs that are outside the 5-minute window
        while time_col[i] - time_col[len(ibi_array) - 1] > window_size_seconds:
            ibi_array.pop(0)

        # Compute HRV index using the current 5-minute IBI array
        hrv_index = compute_hrv_index(ibi_array)

        # Append the HRV index to the result array
        hrv_indices.append((time_col[i], hrv_index))

    return np.array(hrv_indices)

# Example usage:
input_file_path = 'p1.txt'
output_hrv_file_path = 'p1_hrv_indices.txt'

hrv_indices = process_ibis(input_file_path)

# Save the results to a file
np.savetxt(output_hrv_file_path, hrv_indices, fmt='%.6f', delimiter='\t')
```
